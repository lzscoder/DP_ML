简单线性回归的实现

```python
import numpy as np
def fitSLR(X,y):
    n=len(X)
    dinominator = 0
    numerator = 0
    for i in range(0,n):
        numerator += (X[i] - np.mean(X))*(y[i] - np.mean(y))
        dinominator += (X[i] - np.mean(X))**2
    b1 = numerator/float(dinominator)
    b0 = np.mean(y)- b1*float(np.mean(X))
    return b0,b1
def predict(x,b0,b1):
    return b0 + b1*x
```

```python
X = [1,3,2,1,3]
y = [14,24,18,17,27]
b0,b1 = fitSLR(X,y)
print("intercept:",b0,"slope:",b1)
```

```python
X_test = 6
y_test = predict(X_test,b0,b1)
print("y_test:",y_test)
```

多元线性回归,由于矩阵运算的复杂，我们直接使用Sklearn包

```python
from numpy import genfromtxt
from sklearn import datasets, linear_model
dataPath = "Delivery.csv"
deliveryData = genfromtxt(dataPath, delimiter=',')
print ("data:\n",deliveryData)
X = deliveryData[:, :-1]
y = deliveryData[:, -1]
print("X:\n",X,"\ny:\n",y)
```

```python
regr =linear_model.LinearRegression()
regr.fit(X,y)
print("coefficients:",regr.coef_)
print("intercept:",regr.intercept_)
```

```python
xPred = [[102,6]]
yPred = regr.predict(xPred)
print("yPred:",yPred)
```
